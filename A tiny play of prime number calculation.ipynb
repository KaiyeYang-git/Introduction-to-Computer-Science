```python
#input a single argument, and that generates a list of prime numbers 
import numpy as np, pandas as pd
import math,timei,matplotlib

range_input = int(input("Enter your single argument ："))

def MySingle_prime(range_input):
    prime_list = []
    if range_input <= 1:
        return None 
    elif range_input == 2:
        prime_list.append(2)
        return prime_list
    else :
        for i in range(2, range_input + 1):
                for j in range(2, int(math.sqrt(i)) + 1):
                    if i % j == 0:
                        break
                else:
                    prime_list.append(i)
    return prime_list
    
prime_list = MySingle_prime(range_input)
print("The list of prime numbers with a single argument is:",prime_list)

#input a single argument, and that generates a list of prime numbers, using the Sieve of Eratosthenes
import math
import timeit
range_SEinput = int(input("Enter your single argument：)"))

def MySingle_SEprime(range_SEinput):
    prime_SElist = []
    if range_SEinput <= 1:
        return None
    elif range_SEinput == 2:
        prime_SElist.append(2)
        return prime_SElist
    else:
        list_range = [1 for i in range(range_SEinput+1)]
        list_range[0] = 0
        list_range[1] = 0
        for n in range(2,int(math.sqrt(range_SEinput))+1):
            if list_range[n] == 1:
                for j in range(2*n, range_SEinput+1, n):
                    if j > range_SEinput:
                        break
                    list_range[j] = 0
        result_prime = []
        for i in range(len(list_range)):
            if list_range[i] ==1:
                result_prime.append(i)
        return result_prime

result_prime = MySingle_SEprime(range_SEinput)
print("result", result_prime)


#Compare the running times of subtopic a in first method with subtopic a in second method for different list sizes
import matplotlib.pyplot as plt
y1 = []
y2 = []
x = []
for i in range(2,10003,50):
    y1.append(timeit.timeit('MySingle_prime(i)',globals=globals(),number=100))
    y2.append(timeit.timeit('MySingle_SEprime(i)',globals=globals(),number=100))
    x.append(i)
plt.plot(x,y1,label="Q3-a",color ="#F08080")
plt.plot(x,y2,label="Q4-a",color ="#DB7093")
plt.xlabel('Size_input')
plt.ylabel('Time, Unit: s/per 100 time')
plt.legend()
plt.show()
```

```python
#input two arguments, and that generates a list of prime numbers 

start_input = int(input("Enter your first argument："))
end_input = int(input("Enter your second argument："))
def Mydouble_prime(start_input, end_input):
    primeDouble_list = []
    if start_input > end_input or end_input < 2:
        return None
    elif start_input < 2:
        #Prime numbers need to start at 2
        start_input = 2
    for m in range (start_input, end_input + 1):
        for n in range(2, int(math.sqrt(m)) + 1):
            if m % n == 0:
                break
        else:
            primeDouble_list.append(m)
    return primeDouble_list
primeDouble_list = Mydouble_prime(start_input, end_input)
print("The list of prime numbers with two arguments is:",primeDouble_list)

#Qinput two arguments, and that generates a list of prime numbers, using the Sieve of Eratosthenes 

range2_SEstart = int(input("Enter your first argument："))
range2_SEend = int(input("Enter your second argument："))
def Mydouble_SEprime(range2_SEstart, range2_SEend):
    list_range = [1 for i in range(range2_SEend+1)]
    list_range[0] = 0
    list_range[1] = 0
    for n in range(2,int(math.sqrt(range2_SEend))+1):
        if list_range[n] == 1:
            for j in range(2*n, range2_SEend+1, n):
                if j > range2_SEend:
                    break
                list_range[j] = 0
    result_prime = [i for i in range(len(list_range)) if list_range[i]==1]
    result_prime = [i for i in result_prime if i >= range2_SEstart]
    return result_prime

    
prime2_SElist = Mydouble_SEprime(range2_SEstart, range2_SEend)
print("The list of prime numbers with a single argument is:", prime2_SElist)



#Compare the running times of subtopic a in method 1 with subtopic a in method 2 for different list sizes
import matplotlib.pyplot as plt
g1 = []
g2 = []
b = []
for n in range(0,10001,50):
    g1.append(timeit.timeit('Mydouble_prime(n,n+10)',globals=globals(),number=100))
    g2.append(timeit.timeit('Mydouble_SEprime(n,n+10)',globals=globals(),number=100))
    b.append(n)
plt.plot(b,g1,label="M1-b",color ="#F08080")
plt.plot(b,g2,label="M2-b",color ="#DB7093")
plt.xlabel('Size_input')
plt.ylabel('Time, Unit: s/per 100 times')
plt.legend()
plt.show()
```
